### Imports

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import csv
from csv import reader
import time
import winsound

duration = 1000  # milliseconds
freq = 440  # Hz
none_string = 'NaN'

path = 'C:/Users/Mazhar/Projects_Jupyter/6.Scraping-Carrefour/CSV_Pics_Files/Carrefour_Baby_Products_Links.csv'
path_1 = "C:/Users/Mazhar/Projects_Jupyter/6.Scraping-Carrefour/CSV_Pics_Files/Carrefour_Baby_Products.csv"
path_2 = "C:/Users/Mazhar/Projects_Jupyter/6.Scraping-Carrefour/CSV_Pics_Files/Carrefour_Products_NotFound.csv"
executable_path = "C:/Drivers/chromedriver.exe"
```

### Working with Product URLs

```python
# Specify the web page
#raw_url = "https://www.carrefour.pk/mafpak/en/root-maf-category//groceries/cooking-ingredients/cooking-oils-ghee/frying-oil/dalda-cooking-oil-pouch-1ltrx5/p/36445"
```

```python
#----- Creating CSV file for saving scraped data

csv_file = open(path_1, 'w')
csv_writer = csv.writer(csv_file)
#defining headers
csv_writer.writerow(['Product Titles', 'Category_1_Full', 'Category_2', 'Category_3', 'Category_4',  'Price (PKR)',\
                             'Description_1','Description_2','Description_3','Pic_1', 'Pic_2', 'Pic_3', 'Product_Link'])

#----- Creating CSV file for saving Products Data Not Found

csv_notfound = open(path_2, 'w')
csv_writerNotFound = csv.writer(csv_notfound)
#defining headers
csv_writerNotFound.writerow(['Product Data Not Found'])


#----- open Products Link File in read mode

with open(path, 'r') as f:
    
    read_links = reader(f)
    counter = 1
    
    for link in read_links:
        
        #Download webdriver from https://chromedriver.chromium.org/downloads or 
        #https://github.com/mozilla/geckodriver/releases and provide your own path 

        driver = webdriver.Chrome(executable_path) 
        link = str(link[0])
        driver.get(link)
        #print(driver.title)
        driver.minimize_window()
        time.sleep(6)
        
        
        try:
            category_1_full = driver.find_element_by_xpath('//*[@id="__next"]/main/div[1]/div[1]/div')\
                                                            .text.replace("Home","").replace("Groceries", "Groceries, ")
            #print("Category_1_full\n", category_1_full)

        except Exception as e:
            category_1_full = none_string


        try:
            category_2 = driver.find_element_by_xpath('//*[@id="__next"]/main/div[1]/div[1]/div/div[5]').text
            #print("Category_2\n", category_2)

        except Exception as e:
            category_2 = none_string


        try:
            category_3 = driver.find_element_by_xpath('//*[@id="__next"]/main/div[1]/div[1]/div/div[7]').text
            #print("Category_3\n", category_3)

        except Exception as e:
            category_3 = none_string


        try:
            category_4 = driver.find_element_by_xpath('//*[@id="__next"]/main/div[1]/div[1]/div/div[9]').text
            #print("Category_4\n", category_4)

        except Exception as e:
            category_4 = none_string

            
        try:
            title = driver.find_element_by_xpath('/html/body/div/main/div[1]/div[2]/div[2]/section/h1').text
            #print("Title\n", title)            

        except Exception as e:
            title = none_string


        try:    
            price = driver.find_element_by_xpath('/html/body/div/main/div[1]/div[2]/div[2]/section/div/h2')\
                                                                                                .text.replace(" PKR","")
            #print("Price\n", price)              

        except Exception as e:
            price = none_string


        try:
            description_all = driver.find_elements_by_class_name('css-6fdidx')
            description_1 = description_all[0].text
            #print(description_1,"\n")
            description_2 = description_all[1].text
            #print(description_2,"\n")
            description_3 = description_all[2].text
            #print(description_3,"\n")
            
            #check  
            #len(description_all)
            # description_1, description_2, description_3 = description_all
            # print(description_1, "\n", description_2, "\n", description_3)

        except Exception as e:
            description_1 = none_string
            description_2 = none_string
            description_3 = none_string

      
        try:
            images = driver.find_elements_by_class_name('swiper-wrapper')
            img = images[0].find_elements_by_tag_name('img')
            pics = [i.get_attribute('src') for i in img]
            #print(pics)
    
            if len(pics) == 1:
                pics = [pics[0], none_string, none_string]
            elif len(pics) == 2:
                pics = [pics[0], pics[1], none_string]
            elif len(pics) >= 3:
                pics = [pics[0], pics[1], pics[2]]
            else:
                pics = [none_string, none_string, none_string]

        except Exception as e:
            pics = [none_string, none_string, none_string]  

        
        # Writing to CSV file
        try:
            csv_writer.writerow([title, category_1_full, category_2, category_3, category_4, price, 
                         description_1, description_2, description_3, pics[0], pics[1], pics[2], link])
        
        except Exception as e:
            csv_writer.writerow([none_string, none_string, none_string, none_string, none_string, none_string, 
                         none_string, none_string, none_string, none_string, none_string, none_string, link])
            csv_writerNotFound.writerow([link])
            
        
        #time.sleep(5)
        driver.close()
        print("\r Last Completed Product Link:", counter, "|", "Product Link", link, end = '')
        counter += 1
        

        
csv_file.close()
csv_notfound.close()
#winsound.Beep(freq, duration)
print("\n\nTask Completed")
```
